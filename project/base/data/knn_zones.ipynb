```python
%%time

import pandas as pd

# import training csv data
df = pd.read_csv("data.csv")

# df.head(5)


def id_to_int(i):
    return ord(i)


df['b1'] = df['b1'].apply(id_to_int)
df['b2'] = df['b2'].apply(id_to_int)
df['b3'] = df['b3'].apply(id_to_int)


X = df.drop(columns=['zone'])
y = df['zone']


df
```

```python
%%time
# no need to process, data is clean as bean
from sklearn.model_selection import train_test_split
# train test split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
print("shapes:", X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
%%time

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score

knn =  KNeighborsClassifier(leaf_size=1, n_neighbors=8, p=2)
knn.fit(X,y)


#Predict test data set.
y_pred = knn.predict(X_test)#Checking performance our model with classification report.
print(classification_report(y_test, y_pred))#Checking performance our model with ROC Score.
# roc_auc_score(y_test, y_pred)
```
